```python
def unit_vector(v):
    u = vector(map(lambda x: QQ(x.n()), v));
    return u/sqrt(u*u);
###    
def sphere_config(dim, lst):
    it = [iter(lst)]*dim;
    q_config = [unit_vector(v) for v in zip(*it)];
    theta = arccos(max([RIF(x*y) for x in q_config for y in q_config if x!=y]));
    r = 1/2*arccos(1/(1-cos(theta))-1);
    return (r, q_config);
```
